```python
# Importing dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import statsmodels.api as sm
from sqlalchemy import create_engine, Column, Integer, String, Float, Date
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
import psycopg2
import datetime as dt
```

```python
# Load base dataframes from csv files
copperCompanies = pd.read_csv('Resources/copper_company_production.csv')
copperPrices = pd.read_csv('Resources/copper_prices.csv')
gdpPerCapita = pd.read_csv('Resources/gdp_percapita.csv')
goldCompanies = pd.read_csv('Resources/gold_company_production.csv')
goldPrices = pd.read_csv('Resources/gold_prices.csv')
industrialGrowth = pd.read_csv('Resources/industrial_growth.csv')
popGrowth = pd.read_csv('Resources/pop_growth.csv')
copperCountries = pd.read_csv('Resources/top_countries_copper.csv')
goldCountries = pd.read_csv('Resources/top_countries_gold.csv')
```

```python
# Definining tables for SQL
class annual_prices(Base):
    __tablename__ = 'annual_prices'
    __table_args__ = {'extend_existing': True}
    Year = Column(Integer, primary_key=True)
    avg_gold_price = Column('Average Gold Price (per oz t)',Float)
    avg_copper_price = Column('Average Copper Price (per ton)',Float)
    
class monthly_prices(Base):
    __tablename__='monthly_prices'
    __table_args__={'extend_existing':True}
    Month = Column(Date, primary_key=True)
    gold_price = Column('Gold Price (per oz t)',Float)
    copper_price = Column('Copper Price (per ton)',Float)
    
class copper_countries(Base):
    __tablename__='copper_country_production'
    __table_args__={'extend_existing':True}
    year = Column(Integer, primary_key=True)
    United_States = Column('United States',Integer)
    Australia = Column(Integer)
    Chile = Column(Integer)
    China = Column(Integer)
    Congo = Column('Congo (Kinshasa)',Integer)
    Indonesia = Column(Integer)
    Kazakhstan = Column(Integer)
    Mexico = Column(Integer)
    Peru = Column(Integer)
    Russia = Column(Integer)
    Zambia = Column(Integer)
    Other_Countries = Column('Other Countries',Integer)
    World_Total = Column("World Total Production ('000s tons)",Integer)

class gold_countries(Base):
    __tablename__='gold_country_production'
    __table_args__={'extend_existing':True}
    year = Column(Integer, primary_key=True)
    United_States = Column('United States',Integer)
    Argentina = Column(Integer)
    Australia = Column(Integer)
    Brazil = Column(Integer)
    Canada = Column(Integer)
    China = Column(Integer)
    Ghana = Column(Integer)
    Indonesia = Column(Integer)
    Kazakhstan = Column(Integer)
    Mexico = Column(Integer)
    Papua_New_Guinea = Column('Papua New Guinea',Integer)
    Peru = Column(Integer)
    Russia = Column(Integer)
    South_Africa = Column('South Africa',Integer)
    Uzbekistan = Column(Integer)
    Other_Countries = Column('Other Countries',Integer)
    World_Total = Column("World Total Production ('000s oz t)",Integer)
    
class gdp(Base):
    __tablename__='gdp_per_capita'
    __table_args__={'extend_existing':True}
    year = Column(Integer, primary_key=True)
    Argentina = Column(Integer)
    Australia = Column(Integer)
    Brazil = Column(Integer)
    Canada = Column(Integer)
    Chile = Column(Integer)
    China = Column(Integer)
    Congo = Column('DRC',Integer)
    Ghana = Column(Integer)
    Indonesia = Column(Integer)
    Kazakhstan = Column(Integer)
    Mexico = Column(Integer)
    Papua_New_Guinea = Column('Papua New Guinea',Integer)
    Peru = Column(Integer)
    Russia = Column(Integer)
    South_Africa = Column('South Africa',Integer)
    United_States = Column('United States',Integer)
    Uzbekistan = Column(Integer)
    Zambia = Column(Integer)
    World = Column(Integer)
    
class industrial_growth(Base):
    __tablename__='industrial_growth_rate'
    __table_args__={'extend_existing':True}
    year = Column(Integer, primary_key=True)
    Argentina = Column(Float)
    Australia = Column(Float)
    Brazil = Column(Float)
    Canada = Column(Float)
    Chile = Column(Float)
    China = Column(Float)
    Congo = Column('DRC',Float)
    Ghana = Column(Integer)
    Indonesia = Column(Float)
    Kazakhstan = Column(Float)
    Mexico = Column(Float)
    Papua_New_Guinea = Column('Papua New Guinea',Float)
    Peru = Column(Float)
    Russia = Column(Float)
    South_Africa = Column('South Africa',Float)
    United_States = Column('United States',Float)
    Uzbekistan = Column(Float)
    Zambia = Column(Float)
    World = Column(Float)
    
class population_growth(Base):
    __tablename__='population_growth_rate'
    __table_args__={'extend_existing':True}
    year = Column(Integer, primary_key=True)
    Argentina = Column(Float)
    Australia = Column(Float)
    Brazil = Column(Float)
    Canada = Column(Float)
    Chile = Column(Float)
    China = Column(Float)
    Congo = Column('DRC',Float)
    Ghana = Column(Integer)
    Indonesia = Column(Float)
    Kazakhstan = Column(Float)
    Mexico = Column(Float)
    Papua_New_Guinea = Column('Papua New Guinea',Float)
    Peru = Column(Float)
    Russia = Column(Float)
    South_Africa = Column('South Africa',Float)
    United_States = Column('United States',Float)
    Uzbekistan = Column(Float)
    Zambia = Column(Float)
    World = Column(Float)
    
class copper_companies(Base):
    __tablename__='copper_company_production'
    __table_args__={'extend_existing':True}
    year = Column(Integer, primary_key=True)
    Freeport = Column('Freeport-McMoRan',Integer)
    BHP = Column(Integer)
    Southern_Copper = Column('Southern Copper',Integer)
    Glencore = Column(Integer)
    Vale = Column(Integer)
    Rio_Tinto = Column('Rio Tinto',Integer)

class gold_companies(Base):
    __tablename__='gold_company_production'
    __table_args__={'extend_existing':True}
    year = Column(Integer, primary_key=True)
    AngloGold = Column('AngloGold Ashanti',Integer)
    Freeport = Column('Freeport-McMoRan',Integer)
    Goldcorp = Column(Integer)
    Newmont = Column(Integer)
    Barrick = Column(Integer)
    Kinross = Column('Kinross Gold',Integer)
    Newcrest = Column(Integer)
    Polyus = Column(Integer)
```

```python
# Perform statistics checks and graphing prior to using js
# Running tests on price vs production and indices
annual_prices = annual_prices.rename(columns={'Year':'year'})
price_gdp = annual_prices.merge(gdpPerCapita,on='year')
price_gdp = price_gdp[price_gdp['World'].notna()]
price_industrial = annual_prices.merge(industrialGrowth,on='year')
price_industrial = price_industrial[price_industrial['World'].notna()]
price_pop = annual_prices.merge(popGrowth,on='year')
price_pop = price_pop[price_pop['World'].notna()]
price_goldProd = annual_prices.merge(goldCountries,on='year')
price_copperProd = annual_prices.merge(copperCountries,on='year')
```

```python
# Setting graph variables
gdpYear = price_gdp['year']
industrialYear = price_industrial['year']
popYear = price_pop['year']
goldYear = price_goldProd['year']
copperYear = price_copperProd['year']
gdp = price_gdp['World']
industrial = price_industrial['World']
population = price_pop['World']
goldProd = price_goldProd["World Total Production ('000s oz t)"]
copperProd = price_copperProd["World Total Production ('000s tons)"]
gdpPrice_au = price_gdp["Average Gold Price (per oz t)"]
gdpPrice_cu = price_gdp['Average Copper Price (per ton)']
industrialPrice_au = price_industrial['Average Gold Price (per oz t)']
industrialPrice_cu = price_industrial['Average Copper Price (per ton)']
popPrice_au = price_pop['Average Gold Price (per oz t)']
popPrice_cu = price_pop['Average Copper Price (per ton)']
goldPrice = price_goldProd['Average Gold Price (per oz t)']
copperPrice = price_copperProd['Average Copper Price (per ton)']
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3)
ax1.plot(gdpYear,gdp)
ax2.plot(gdpYear,gdpPrice_au)
ax3.plot(gdpYear,gdpPrice_cu)
fig.suptitle('World GDP (top), Gold Price (mid), and Copper Price (bottom)')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3)
ax1.plot(industrialYear,industrial)
ax2.plot(industrialYear,industrialPrice_au)
ax3.plot(industrialYear,industrialPrice_cu)
fig.suptitle('World Industrial Growth Rate (top), Gold Price (mid), and Copper Price (bottom)')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3)
ax1.plot(popYear,population)
ax2.plot(popYear,popPrice_au)
ax3.plot(popYear,popPrice_cu)
fig.suptitle('World Population Growth Rate (top), Gold Price (mid), and Copper Price (bottom)')
```

```python
fig, (ax1,ax2) = plt.subplots(2)
ax1.plot(goldYear,goldProd)
ax2.plot(goldYear,goldPrice)
fig.suptitle('World Gold Production (top), Gold Price (bottom')
```

```python
fig, (ax1,ax2) = plt.subplots(2)
ax1.plot(copperYear,copperProd)
ax2.plot(copperYear,copperPrice)
fig.suptitle('World Copper Production (top), Copper Price (bottom')
```

It appears that Gold and Copper production dont necessarily correlate with price, but rather the mining companies produce as much as they can every year. Gold may have had a correlation between production and price up until 2008 when production was decreasing steadily while price increased, however production has been increase every year since then, even with a dip in price. This may be due to Gold demand in the electronics industry

The indices show a better story. It appears that Gold and Copper prices may both correlate positively with world GDP and Industrial Growth Rate, and negatively with Population Growth Rate. It should make sense for the first two, whereas countries have more money and are growing more in industry will cause more demand for Gold and Copper due to electronics and industrial uses. I would expect the industrial growth rate to have continued on an upward trend for the period this data represents, however the economic recession in 2008 is likely the culprit in the sharp decrease in industrial growth, as well as Copper price (due to a lower demand while production continued to increase)

```python
# Scatter plots to verify any correlation
# GDP
slope_au, v_int_au, r_au, p_au, std_err_au = sts.linregress(gdp,gdpPrice_au)
fit_au = slope_au*gdp+v_int_au
slope_cu, v_int_cu, r_cu, p_cu, std_err_cu = sts.linregress(gdp,gdpPrice_cu)
fit_cu = slope_cu*gdp+v_int_cu
fig, (ax1,ax2) = plt.subplots(2)
ax1.scatter(gdp,gdpPrice_au)
ax1.plot(gdp,fit_au,'--')
ax2.scatter(gdp,gdpPrice_cu)
ax2.plot(gdp,fit_cu,'--')
fig.suptitle('World GDP vs Gold Price (top), vs Copper Price (bottom)')
ax1.grid(b=True,axis='both')
ax2.grid(b=True,axis='both')
plt.show()
r_au = round(r_au,3)
r_cu = round(r_cu,3)
slope_au = round(slope_au,4)
slope_cu = round(slope_cu,4)
v_int_au = round(v_int_au,2)
v_int_cu = round(v_int_cu,2)
print(f'Correlation Coefficient for Gold Price = {r_au}')
print(f'Correlation Coefficient for Copper Price = {r_cu}')
print(f'Gold model: Gold Price = {slope_au} * World GDP + {v_int_au}')
print(f'Copper model: Copper Price = {slope_cu} * World GDP + {v_int_cu}')
```

```python
# Industrial Growth Rate
slope_au, v_int_au, r_au, p_au, std_err_au = sts.linregress(industrial,industrialPrice_au)
fit_au = slope_au*industrial+v_int_au
slope_cu, v_int_cu, r_cu, p_cu, std_err_cu = sts.linregress(industrial,industrialPrice_cu)
fit_cu = slope_cu*industrial+v_int_cu
fig, (ax1,ax2) = plt.subplots(2)
ax1.scatter(industrial,industrialPrice_au)
ax1.plot(industrial,fit_au,'--')
ax2.scatter(industrial,industrialPrice_cu)
ax2.plot(industrial,fit_cu,'--')
fig.suptitle('World Industrial Growth Rate vs Gold Price (top), vs Copper Price (bottom)')
ax1.grid(b=True,axis='both')
ax2.grid(b=True,axis='both')
plt.show()
r_au = round(r_au,3)
r_cu = round(r_cu,3)
slope_au = round(slope_au,2)
slope_cu = round(slope_cu,2)
v_int_au = round(v_int_au,2)
v_int_cu = round(v_int_cu,2)
print(f'Correlation Coefficient for Gold Price = {r_au}')
print(f'Correlation Coefficient for Copper Price = {r_cu}')
```

```python
# Population Growth Rate
slope_au, v_int_au, r_au, p_au, std_err_au = sts.linregress(population,popPrice_au)
fit_au = slope_au*population+v_int_au
slope_cu, v_int_cu, r_cu, p_cu, std_err_cu = sts.linregress(population,popPrice_cu)
fit_cu = slope_cu*population+v_int_cu
fig, (ax1,ax2) = plt.subplots(2)
ax1.scatter(population,popPrice_au)
ax1.plot(population,fit_au,'--')
ax2.scatter(population,popPrice_cu)
ax2.plot(population,fit_cu,'--')
fig.suptitle('World Population Growth Rate vs Gold Price (top), vs Copper Price (bottom)')
ax1.grid(b=True,axis='both')
ax2.grid(b=True,axis='both')
plt.show()
r_au = round(r_au,3)
r_cu = round(r_cu,3)
slope_au = round(slope_au,2)
slope_cu = round(slope_cu,2)
v_int_au = round(v_int_au,2)
v_int_cu = round(v_int_cu,2)
print(f'Correlation Coefficient for Gold Price = {r_au}')
print(f'Correlation Coefficient for Copper Price = {r_cu}')
```

Population Growth Rate and GDP have decent correlation coefficients and will be looked into further

```python
# Productions
slope, v_int, r, p, std_err = sts.linregress(goldProd,goldPrice)
fit = slope*goldProd+v_int
plt.scatter(goldProd,goldPrice)
plt.plot(goldProd,fit,'--')
plt.suptitle('World Gold Production vs Gold Price')
plt.grid(b=True,axis='both')
plt.show()
r = round(r,3)
slope = round(slope,2)
v_int = round(v_int,2)
print(f'Correlation Coefficient for Gold Price = {r}')
```

```python
# Productions
slope, v_int, r, p, std_err = sts.linregress(copperProd,copperPrice)
fit = slope*copperProd+v_int
plt.scatter(copperProd,copperPrice)
plt.plot(copperProd,fit,'--')
plt.suptitle('World Copper Production vs Copper Price')
plt.grid(b=True,axis='both')
plt.show()
r = round(r,3)
slope = round(slope,2)
v_int = round(v_int,2)
print(f'Correlation Coefficient for copper Price = {r}')
```

Based on the correlations, even though the production vs price for both Gold and Copper are over , visually the data does not look like it trends very well. Instead, multi-variate regression analysis will be done with Population Growth Rate and GDP per Capita

```python
# Combine GDP and Pop Growth DFs
gdp_formerge = price_gdp[['year','Average Gold Price (per oz t)','Average Copper Price (per ton)','World']]
pop_formerge = price_pop[['year','World']]
gdp_pop = gdp_formerge.merge(pop_formerge,on='year')
gdp_pop = gdp_pop.rename(columns={'World_x':'World GDP','World_y':'World Population Growth Rate'})
gdp_pop
```

```python
# reg_df = gdp_pop.merge()
goldProd_formerge = price_goldProd[['year',"World Total Production ('000s oz t)"]]
copperProd_formerge = price_copperProd[['year',"World Total Production ('000s tons)"]]
```

```python
reg_df = gdp_pop.merge(goldProd_formerge,on='year')
reg_df = reg_df.merge(copperProd_formerge,on='year')
reg_df
```

```python
# Run multiple linear model tests using GDP, Population Growth, and World Production as variables
from sklearn import linear_model
x = reg_df[['World GDP','World Population Growth Rate']]
y = reg_df['Average Gold Price (per oz t)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Gold with Wolrd GDP and Population Growth R-squared = {lm.score(x,y)}')
x = reg_df[['World GDP',"World Total Production ('000s oz t)"]]
y = reg_df['Average Gold Price (per oz t)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Gold with Wolrd GDP and World Production R-squared = {lm.score(x,y)}')
x = reg_df[['World Population Growth Rate',"World Total Production ('000s oz t)"]]
y = reg_df['Average Gold Price (per oz t)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Gold with Wolrd Population Growth and World Production R-squared = {lm.score(x,y)}')
x = reg_df[['World GDP','World Population Growth Rate',"World Total Production ('000s oz t)"]]
y = reg_df['Average Gold Price (per oz t)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Gold with all 3 variables R-squared = {lm.score(x,y)}')
```

```python
x = reg_df[['World GDP','World Population Growth Rate']]
y = reg_df['Average Copper Price (per ton)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Copper with Wolrd GDP and Population Growth R-squared = {lm.score(x,y)}')
x = reg_df[['World GDP',"World Total Production ('000s tons)"]]
y = reg_df['Average Copper Price (per ton)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Copper with Wolrd GDP and World Production R-squared = {lm.score(x,y)}')
x = reg_df[['World Population Growth Rate',"World Total Production ('000s tons)"]]
y = reg_df['Average Copper Price (per ton)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Copper with Wolrd Population Growth and World Production R-squared = {lm.score(x,y)}')
x = reg_df[['World GDP','World Population Growth Rate',"World Total Production ('000s tons)"]]
y = reg_df['Average Copper Price (per ton)']
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
print(f'Copper with all 3 variables R-squared = {lm.score(x,y)}')
```

Best common regression is simply to World GDP \
Gold model: Gold Price = 0.1221 * World GDP + -491.51\
Copper model: Copper Price = 0.4841 * World GDP + -315.6

```python
reg_df['Gold Model'] = 0.1221*reg_df['World GDP']-491.51
reg_df['Copper Model'] = 0.4841*reg_df['World GDP']-315.6
```

```python
plt.plot(reg_df['year'],reg_df['Average Gold Price (per oz t)'])
plt.plot(reg_df['year'],reg_df['Gold Model'])
plt.suptitle('Gold Actual vs Model Price')
plt.show()
```

```python
plt.plot(reg_df['year'],reg_df['Average Copper Price (per ton)'])
plt.plot(reg_df['year'],reg_df['Copper Model'])
plt.suptitle('Copper Actual vs Model Price')
plt.show()
```
